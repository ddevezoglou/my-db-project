```python
import pandas as pd
from db.connection import get_engine
```

```python
engine=get_engine()
```

```python
users_df=pd.read_sql("select * from app.users u",engine)
```

```python
projects_df=pd.read_sql("select * from app.projects p",engine)
```

```python
projects_df.head()
```

```python
merged_df=users_df.merge(projects_df,left_on="id",right_on="user_id",how="inner")
```

```python
results_df=(merged_df.groupby(["id_x","name","surname"])["budget"].sum().reset_index())
```

```python
results_df
```

```python
results_df=results_df.sort_values(by="budget",ascending=False)
```

```python
results_df.to_sql("users_total_budget",engine,schema="app",if_exists="replace",index=False)
```
