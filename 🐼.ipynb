```python
def calculate_user_total_budget(engine):
    users_df = pd.read_sql("SELECT * FROM app.users;", engine)
    projects_df = pd.read_sql("SELECT * FROM app.projects;", engine)

    merged_df = users_df.merge(
        projects_df,
        left_on="id",
        right_on="user_id"
    )

    result_df = (
        merged_df
        .groupby(["id_x", "name", "surname"])["budget"]
        .sum()
        .reset_index()
    )

    result_df = result_df.sort_values(by="budget", ascending=False)

    return result_df
```

```python
result_df=calculate_user_total_budget(engine)
result_df
```
